# kNN

一般kNN的流程;

(1) 计算已知类别数据集中的点与当前点之间的距离;\

(2) 按照距离递增次序排序;\

(3) 选取与当前点距离最小的k个点;\

(4) 确定前k个点所在类别的出现频率;\

(5) 返回前k个点出现频率最高的类别作为当前点的预测分类。\

更多距离的度量可以在https://my.oschina.net/hunglish/blog/787596 找到

```python
%matplotlib inline
import os
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from collections import Counter
```

```python
file_path = './datingTestSet.txt'
raw_data = []
labels = []
with open(file_path, 'r') as fp:
    for line in fp.readlines():
        line = line.strip()
        elems = line.split('\t')
        like = elems[3]
        label = 1 if like == 'largeDoses' else 2 if like == 'smallDoses' else 3
        raw_data.append([int(elems[0]), float(elems[1]), float(elems[2])])
        labels.append(label)
        
raw_data = np.array(raw_data)
labels = np.array(labels).reshape(-1,1)
```

```python
print(raw_data.shape)
colors = np.concatenate((labels/3, labels/6, np.zeros_like(labels)),1)
print(colors.shape)
plt.scatter(raw_data[:,0], raw_data[:,1], c=colors)
```

## 数据的归一化

由于数据具有不同的单位，因此需要归一化来去除量纲

```python
def normalize(raw_data):
    min_data = np.min(raw_data, 0)
    max_data = np.max(raw_data, 0)
    
    epsilon = np.ones_like(max_data)*1e-6
    ranges = ((max_data-min_data)+epsilon)
    norm_data = (raw_data-min_data)/ranges
    
    return norm_data, max_data, min_data

norm_data, max_data, min_data = normalize(raw_data)
print(">>> raw\n", raw_data[:10,:])
print(">>> norm\n", norm_data[:10,:])
```

## 最近邻算法

对每个点都计算距离，然后选择k个距离最近的类别作为该点的类别

```python
def classfy0(sample, norm_data, labels, k):
    diff = sample - norm_data
    ''' 使用L2范数 '''
    scores = np.linalg.norm(diff, axis=1)
    arg_index = np.argsort(scores)[k]
    counter = Counter(labels[arg_index])
    max_label = 0
    max_cnt = 0
    for label, num in counter.items():
        if num > max_cnt:
            max_label = label
            max_cnt = num
    return max_label
```

## 交叉验证的方法

1. 产生下标数组
2. 将数组乱序，抽取前80%作为训练集，后20%作为验证集
3. 从1到10到增加k，记录下最后的准确率

```python
import random

N = norm_data.shape[0]
index = list(range(N))
random.shuffle(index)
train_index = index[:int(N*0.8)]
test_index  = index[int(N*0.8)+1:]
train_set, train_label = norm_data[train_index,:], labels[train_index]
test_set, test_label = norm_data[test_index,:], labels[test_index]
```

```python
record = {}
for k in range(2,10):
    currect = 0
    N = len(test_label)
    for sample, label in zip(test_set, test_label):
        res = classfy0(sample, train_set, train_label, k)
#         print(">>> result: {}\tlabel: {}".format(res, label))
        if (res == label):
            currect += 1
    record[k] = currect/N
```

## 对手写数字进行kNN分类

### 读取训练集和验证集

训练集和验证集都在文件夹中，每一个文件都是一张二值图，样本的标签为文件名的第一个字符

```python
def img2vector(dir_path, filename):
    label = int(filename[0])
    file_path = os.path.join(dir_path, filename)
    vec = np.zeros((1024))
    with open(file_path, 'r') as fp:
        for i in range(32):
            line = fp.readline()
            for j in range(32):
                vec[i*32+j] = int(line[j])
    return vec, label

def load_dataset(dir_path):
    file_names = os.listdir(dir_path)
    features = []
    labels = []
    for file_name in file_names:
        vec, label = img2vector(dir_path, file_name)
        features.append(vec)
        labels.append(label)
    return features, labels
```

```python
train_dir_path = "./trainingDigits/"
test_dir_path  = "./testDigits/"

train_features, train_labels = load_dataset(train_dir_path)
test_features, test_labels = load_dataset(test_dir_path)

train_features = np.array(train_features)
test_features = np.array(test_features)

train_labels = np.array(train_labels).reshape(-1,1)
test_labels  = np.array(test_labels).reshape(-1,1)
```

### 查看load的数据是否是对的

随机从train和test中选择一个数据，之后用matplotlib绘制出来

```python
train_idx = np.random.randint(0, train_features.shape[0])
test_idx = np.random.randint(0, test_features.shape[0])

train_vec = train_features[train_idx, :].reshape(32,-1)
test_vec = test_features[test_idx, :].reshape(32,-1)

train_label = train_labels[train_idx, 0]
test_label = test_labels[test_idx, 0]

fig, axes = plt.subplots(ncols=2)
axes[0].imshow(train_vec)
axes[0].set_title(str(train_label))

axes[1].imshow(test_vec)
axes[1].set_title(str(test_label))
```

### 测试函数

```python
def test_classify(train_data, train_label, test_data, test_label, k):
    N = test_data.shape[0]
    cur_n = 0
    for i in range(N):
        res = classfy0(test_data[i,:], train_data, train_label, k)
        if res == test_label[i, 0]:
            cur_n += 1
    return cur_n/N
```

```python
print("kNN accuracy is {}".format(test_classify(train_features, train_labels, test_features, test_labels, 4)))
```
